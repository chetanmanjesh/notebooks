# Fastest Inferences Runtime Per Sentence In Current  Experimental Set-up

Among similar English NLP-pipelines (i.e. part of the pipeline enabled or disabled): 
* **en_core_web_sm (only ner enabled)** is on an average **59.08x** faster (CPU), **2.33x** faster (GPU) compared to **en_core_web_trf (only ner enabled)**.
* **en_core_web_sm (nothing disabled)** is **30.73x** faster (CPU) and **0.91x** faster (GPU) **en_core_web_trf (nothing disabled)**.

GPUs are on average faster than CPUs by the following margins:
* **2.12x** faster for **en_core_web_sm (only ner enabled)**
* **1.47x** faster for **en_core_web_sm (nothing disabled)**
* **55.31x** faster for **en_core_web_trf (only ner enabled)**
* **39.84x** faster for **en_core_web_trf (nothing disabled)**
